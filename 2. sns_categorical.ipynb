jitter parameter is a boolean that controls whether or not to add jitter to the categorical scatter points. Jitter is a small amount of random noise added to the horizontal or vertical position of each data point. This is particularly useful when you have overlapping points in a categorical scatter plot, as it helps to visualize the density of points more clearly.
________________________________________________

# jitter=True (default)
 a small amount of random noise is added to the x-axis (categorical variable) position of each data point. This is done to prevent overlap and make it easier to see the distribution of points.

 # jitter=False
 no jitter is added, and the points are plotted without any random displacement along the x-axis. This might result in overlapping points if there are multiple data points with the same categorical value.

```python
sns.catplot(data=tips, x="day", y="total_bill", jitter=False)
#sns.catplot(data=tips, x="day", y="total_bill", jitter=True)
```

kind parameter determines the type of plot to create within the categorical plot. The kind parameter accepts several options, each corresponding to a different type of categorical plot.

# kind="strip" (default):

    Creates a strip plot where individual data points are shown along the categorical axis. This is similar to a scatter plot.

# kind="swarm":

    Similar to a strip plot, but points are adjusted so that they do not overlap, providing a better representation of the distribution of points.

# kind="box":

    Creates a box plot that displays the distribution of the data's quartiles, highlighting the median, and identifying potential outliers.

# kind="violin":

    Produces a violin plot that combines aspects of a box plot and a kernel density plot. It shows the distribution of the data and includes information about the median and interquartile range.

# kind="bar":

    Creates a bar plot showing the mean or other summary statistic for each category.

# kind="count":

    Displays the count of observations in each category using bars.

A box plot provides a five-numbered statistical summary, which delivers valuable information for understanding the existing variables. The five number summary comprised of Minimum, First Quartile (Q1), 2nd Quartile (Q2) or median, Third Quartile (Q3) and Maximum. The difference between the Third Quartile (Q3) and First Quartile (Q1) is called the interquartile range (IQR).

Sometimes, a box plot also helps in identifying the outliers that are far away than Q3+(1.5 *IQR) or Q1-(1.5 *IQR).

A violin plot is a method of plotting numeric data and can be understood as a combination of a box plot and a kernel density (KDE) plot. It provides a visualization of the distribution of the data, its probability density, and its cumulative distribution.

```python
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex", kind="violin",
)
```

# Why Use a Violin Plot?

1. Understanding Distribution: Violin plots allow you to visualize the distribution of a numeric variable for one or several groups.
2. Density Estimation: The width of the plot provides a density estimation of the variable at different values.
3. Multiple Groups: Its excellent for comparing the distribution of a variable across multiple groups.
Creating a Violin Plot with Seaborn


bw and cut parameters are specific to the kind="violin" option, which creates violin plots. Let's discuss each of these parameters:

# bw (bandwidth):

    The bw parameter controls the width of the kernel density estimate in the violin plot. It determines the smoothness of the distribution curve within each violin.
    A smaller bw value results in a narrower and more detailed distribution curve, while a larger bw value leads to a smoother and wider curve.
    In your example, bw=.15 sets the bandwidth to 0.15, which means a relatively narrower kernel density estimate.

# cut:

    The cut parameter controls the extent of the violin beyond the data range. It determines how far the violin plot extends beyond the minimum and maximum values of the data.
    A cut value of 0 means that the violin plot precisely follows the data range, while a non-zero cut value extends the violin plot beyond the data range.
    In your example, cut=0 indicates that the violin plot should precisely follow the data range without any extension.

```python
sns.catplot(
    data=tips, x="total_bill", y="day", hue="sex",
    kind="violin", bw=.15, cut=0,
)
```

The split parameter is useful when you have a hue variable, and you want to visually compare the distributions between the different levels of the hue variable.

# split=True
separate violins are drawn for each level of the hue variable. For example, if hue="sex", you would get separate violins for each gender.

# split=False
the violins are combined for each level of the hue variable, resulting in a single, wider violin that represents the combined distribution for both levels of the hue variable.

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", split=True,
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", split=False,
)
```

# inner="stick":

    The inner parameter specifies the type of plot to draw inside the violins. In this case, inner="stick" means that individual data points will be represented as small sticks or lines within the violin plot. This provides a visual indication of the distribution of individual observations.

    Other options for the inner parameter include:
        "quartile" (default): Draws quartiles inside the violins.
        "point": Draws individual data points as points.
        "box": Draws a miniature box plot inside the violins.
        None: Does not draw anything inside the violins.

# palette="pastel":

    The palette parameter sets the color palette for the plot. It defines the colors used for different levels of the hue variable (in this case, "sex").

    In your example, palette="pastel" specifies a pastel color palette. Seaborn provides several built-in palettes, and you can also create custom palettes or use other predefined ones.

    Other options for the palette parameter include:
        A list of colors to use.
        Named seaborn palettes (e.g., "deep", "muted", "bright").
        Matplotlib colormap names.

    You can explore different palettes and choose the one that suits your visualization preferences and enhances readability.

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", inner="stick", split=True, palette="pastel",
)
```

    palette="ch:.25":

        Here, ch:.25 is a shorthand notation for selecting a specific color palette. In Seaborn, the ch: prefix indicates that the color palette should be a qualitative (categorical) palette. The .25 suffix controls the saturation level, where 0 is grayscale, and 1 is full color.

        In this case, the ch:.25 palette is a qualitative palette with reduced saturation (25%). This can be useful to create plots with more subdued or muted colors.

        The ch: prefix stands for "Color Brewer," referring to the Color Brewer color schemes that Seaborn provides.

```python
sns.catplot(data=titanic, x="deck", kind="count", palette="ch:.25")
```

edgecolor parameter is used to set the color of the edges of the bars or elements in the plot. It defines the color of the borders around the individual bars or other graphical elements in the categorical plot.

    edgecolor=".6":

        This sets the edge color to a light gray color with a specified intensity. The value .6 represents the intensity of the gray color, where 0 is black, 1 is white, and values in between represent shades of gray.

        In this case, edgecolor=".6" means that the edges of the bars in the count plot will be light gray.

```python
sns.catplot(
    data=titanic, y="deck", hue="class", kind="count",
    palette="pastel", edgecolor=".6",
)
```

```python
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="point")
```

    palette={"male": "g", "female": "m"}:
        The palette parameter is used to set the color palette for the different levels of the "sex" variable. In this case, it's a dictionary mapping the levels ("male" and "female") to specific colors ("g" for green and "m" for magenta).

    markers=["^", "o"] and linestyles=["-", "--"]:
        The markers parameter specifies the marker styles for each level of the "sex" variable. In this case, "male" is represented by an upward-pointing triangle ("^") and "female" by a circle ("o").
        The linestyles parameter specifies the line styles connecting the markers. "male" is represented by a solid line ("-"), and "female" by a dashed line ("--").

```python
sns.catplot(
    data=titanic, x="class", y="survived", hue="sex",
    palette={"male": "g", "female": "m"},
    markers=["^", "o"], linestyles=["-", "--"],
    kind="point"
)
```

    aspect=.7:

        Here, aspect=.7 sets the aspect ratio to 0.7. This means that each facet (subplot) in the column of the faceted plot will have a height that is 0.7 times its width.

        A value less than 1 makes each subplot wider, while a value greater than 1 makes each subplot taller.

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="smoker",
    kind="swarm", col="time", aspect=.7,
)
```

    data=titanic:
        Specifies the DataFrame (titanic) containing the data for the plot.

    x="fare", y="embark_town", row="class":
        Defines the variables to be used for the x-axis ("fare"), y-axis ("embark_town"), and rows ("class") in the faceted box plot.

    kind="box":
        Specifies the type of plot to create, in this case, a box plot.

    orient="h":
        Sets the orientation of the box plot to horizontal ("h").

    sharex=False:
        Ensures that each subplot (box plot) has its own x-axis, allowing different y-axis scales.

    margin_titles=True:
        Enables the use of margin titles, which are titles placed in the margins of the facet grid.

    height=1.5, aspect=4:
        height=1.5 sets the height of each subplot to 1.5 inches.
        aspect=4 sets the aspect ratio (width to height) of each subplot to 4.

    g.set(xlabel="Fare", ylabel=""):
        Sets the x-axis label to "Fare" and removes the y-axis label (empty string).

    g.set_titles(row_template="{row_name} class"):
        Sets the titles for each row using a template. In this case, it uses the row name and appends " class" to it.

```python
g = sns.catplot(
    data=titanic,
    x="fare", y="embark_town", row="class",
    kind="box", orient="h",
    sharex=False, margin_titles=True,
    height=1.5, aspect=4,
)
g.set(xlabel="Fare", ylabel="")
g.set_titles(row_template="{row_name} class")
```
